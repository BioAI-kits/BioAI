```python
from bioai.data.readData import ReadData
import pandas as pd 
```

```python
datas, label = ReadData(dataFiles=['./example/cnv.csv.gz',
                                   './example/met.csv.gz',
                                   './example/rna.csv.gz'
                                   ], 
                        labelFile='./example/label.csv'
                        ).run()
```

```python
names = []
for df_tmp in datas:
    names += list(set(df_tmp.index.values.tolist()))
names += list(set(label.index.values.tolist()))
df = pd.value_counts(names).to_frame().reset_index()
df.columns = ['id', 'count_']
df = df[df.count_ == len(datas)+1]
```
